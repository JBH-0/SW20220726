# 숫자 야구 게임 총 1000번 수행

```python
import random
count=0 #공격한 횟수
scount=0 #strike count
bcount=0 #ball count
avg=0 #공격 회수 최종 평균
for i in range(1,1001):
  print('\n---',i,'번째---')

  #정답 생성
  cor=list(range(0,3))
  i=0
  while(i<len(cor)):
    cor[i]=str(random.randint(1,9))
    j=0
    while(j<i):
      if cor[i] == cor[j]:
        i-=1
        break
      j+=1
    i+=1
  print('<정답>')
  for i in range(len(cor)):
    print(cor[i],end=' ')
  print()

  #경우의 수가 담겨진 배열 생성
  bb=[] #줄인 경우의 수
  for i in range(10):
    for j in range(10):
      if not(i==j):
        for n in range(10):
          if not(j==n or i==n):
            bb.append(str(i)+str(j)+str(n))
  
  #Strike와 Ball 판별하여 경우의 수 제거하기
  rep=True
  while(rep):
    #공격자 난수 발생
    rand=random.randint(0,len(bb)-1)
    at=bb[rand]
    print('공격:',at)
    count+=1 #공격한 횟수 1 증가

    #s와 b 갯수 구하기
    i=0
    while(i<len(cor)):
      j=0
      while(j<len(cor)):
        if cor[i]==at[j]:#공격 숫자가 정답에 있으면
          if i==j: #위치도 같으면 
            scount+=1 #strike 수 증가
          else: # 공격 숫자가 정답에 있지만, 위치는 다르면
            bcount+=1 #ball 수 증가
        j+=1
      j=0
      i+=1
    print('strike:',scount,', ball:',bcount)
    if scount==0: 
      if bcount==0: #0Strike, 0Ball
        print('0s 0b')
        i=0
        while(i<len(bb)):
          #같은 수가 1개라도 있으면 삭제하라
          if bb[i][0] in at or bb[i][1] in at or bb[i][2] in at: 
            del bb[i]
            i-=1
          i+=1
      elif bcount==1: #0Strike, 1Ball
        print('0s 1b')
        i=0
        while(i<len(bb)):
          #백, 십의 자리는 같은 수가 없고 일의 자리는 자기 위치 제외한 곳에 있으면 continue
          if bb[i][0] not in at and bb[i][1] not in at and (bb[i][2]==at[0] or bb[i][2]==at[1]): 
            i+=1
            continue
           #백, 일의 자리는 같은 수가 없고 십의 자리는 자기 위치 제외한 곳에 있으면 continue
          elif bb[i][0] not in at and bb[i][2] not in at and (bb[i][1]==at[0] or bb[i][1]==at[2]):
            i+=1
            continue
           #십, 일의 자리는 같은 수가 없고 백의 자리는 자기 위치 제외한 곳에 있으면 continue
          elif bb[i][1] not in at and bb[i][2] not in at and (bb[i][0]==at[1] or bb[i][0]==at[2]):
            i+=1
            continue
          else: #그렇지 않으면 삭제하라
            del bb[i]
            i-=1
          i+=1
      elif bcount==2: #0Strike, 2Ball
        print('0s 2b')
        i=0
        while(i<len(bb)):
          #백의 자리는 같은 수가 없고 십,일의 자리는 자기 위치 제외한 곳에 있으면 continue
          if bb[i][0] not in at and (bb[i][1]==at[0] or bb[i][1]==at[2]) and (bb[i][2]==at[0] or bb[i][2]==at[1]):
            i+=1
            continue
          #십의 자리는 같은 수가 없고 백,일의 자리는 자기 위치 제외한 곳에 있으면 continue
          elif (bb[i][0]==at[1] or bb[i][0]==at[2]) and bb[i][1] not in at and (bb[i][2]==at[0] or bb[i][2]==at[1]):
            i+=1
            continue
          #일의 자리는 같은 수가 없고 백,십의 자리는 자기 위치 제외한 곳에 있으면 continue
          elif (bb[i][0]==at[1] or bb[i][0]==at[2]) and (bb[i][1]==at[0] or bb[i][1]==at[2]) and bb[i][2] not in at:
            i+=1
            continue
          else: #그렇지 않으면 삭제하라
            del bb[i]
            i-=1
          i+=1
      elif bcount==3: #0Strike, 3Ball
        print('0s 3b')
        i=0
        while(i<len(bb)):
          #모든 숫자가 존재하고, 자기 위치에 있지 않으면 continue
          if (bb[i][0]==at[1] or bb[i][0]==at[2]) and (bb[i][1]==at[0] or bb[i][1]==at[2]) and (bb[i][2]==at[0] or bb[i][2]==at[1]):
            i+=1
            continue
          else: #그렇지 않으면 삭제하라
            del bb[i]
            i-=1
          i+=1 
    elif scount==1: 
      if bcount==0: #1Strike, 0Ball
        print('1s 0b')
        i=0
        while(i<len(bb)):
          #백의 자리는 같다. 십,일의 자리는 없으면 continue
          if bb[i][0]==at[0] and bb[i][1] not in at and bb[i][2] not in at: 
            i+=1
            continue
          #십의 자리는 같다. 백,일의 자리는 없으면 continue
          elif bb[i][0] not in at and bb[i][1]==at[1] and bb[i][2] not in at:
            i+=1
            continue
          #일의 자리는 같다. 백,십의 자리는 없으면 continue
          elif bb[i][0] not in at and bb[i][1] not in at and bb[i][2]==at[2]:
            i+=1
            continue
          else: #그렇지 않으면 삭제하라
            del bb[i]
            i-=1
          i+=1
      elif bcount==1: #1Strike, 1Ball
        print('1s 1b')
        i=0
        while(i<len(bb)):
          if bb[i][0]==at[0]: #백의 자리는 같다.
            #십의 자리는 없고, 일의 자리는 십의 자리에 위치하면 continue
            if bb[i][1] not in at and bb[i][2]==at[1]: 
              i+=1
              continue
            #일의 자리는 없고, 십의 자리는 일의 자리에 위치하면 continue
            elif bb[i][2] not in at and bb[i][1]==at[2]: 
              i+=1
              continue
            else: #그렇지 않으면 삭제
              del bb[i]
              i-=1
          elif bb[i][1]==at[1]: #십의 자리는 같다.
            #백의 자리는 없고, 일의 자리는 백의 자리에 위치하면 continue
            if bb[i][0] not in at and bb[i][2]==at[0]: 
              i+=1
              continue
            #일의 자리는 없고, 백의 자리는 일의 자리에 위치하면 continue
            elif bb[i][2] not in at and bb[i][0]==at[2]: 
              i+=1
              continue
            else: #그렇지 않으면 삭제
              del bb[i]
              i-=1
          elif bb[i][2]==at[2]: #일의 자리는 같다.
            #백의 자리는 없고, 십의 자리는 백의 자리에 위치하면 continue
            if bb[i][0] not in at and bb[i][1]==at[0]: 
              i+=1
              continue
            #십의 자리는 없고, 백의 자리는 십의 자리에 위치하면 continue
            elif bb[i][1] not in at and bb[i][0]==at[1]: 
              i+=1
              continue
            else: #그렇지 않으면 삭제
              del bb[i]
              i-=1
          else: #그렇지 않으면 삭제
            del bb[i]
            i-=1
          i+=1
      elif bcount==2: #1Strike, 2Ball
        print('1s 2b')
        i=0
        while(i<len(bb)):
          if bb[i][0]==at[0]: #백의 자리는 같다.
            #십의 자리와 일의 자리 위치만 다르면 continue
            if bb[i][1]==at[2] and bb[i][2]==at[1]: 
              i+=1
              continue
            else: #그렇지 않으면 삭제
              del bb[i]
              i-=1
          elif bb[i][1]==at[1]: #십의 자리는 같다.
            #백의 자리와 일의 자리 위치만 다르면 continue
            if bb[i][0]==at[2] and bb[i][2]==at[0]: 
              i+=1
              continue
            else: #그렇지 않으면 삭제
              del bb[i]
              i-=1
          elif bb[i][2]==at[2]: #일의 자리는 같다.
            #백의 자리와 십의 자리 위치만 다르면 continue
            if bb[i][0]==at[1] and bb[i][1]==at[0]: 
              i+=1
              continue
            else: #그렇지 않으면 삭제
              del bb[i]
              i-=1
          else: #그렇지 않으면 삭제
            del bb[i]
            i-=1
          i+=1
    elif scount==2 and bcount==0:  #2Strike, 0Ball
      print('2s 0b')
      i=0
      while(i<len(bb)):
        #백,십의 자리는 같다. 일의 자리는 없으면 continue
        if bb[i][0]==at[0] and bb[i][1]==at[1] and bb[i][2] not in at: 
          i+=1
          continue
        #백,일의 자리는 같다. 십의 자리는 없으면 continue
        elif bb[i][0]==at[0] and bb[i][1] not in at and bb[i][2]==at[2]: 
          i+=1
          continue
        #십,일의 자리는 같다. 백의 자리는 없으면 continue
        elif bb[i][0] not in at and bb[i][1]==at[1] and bb[i][2]==at[2]: 
          i+=1
          continue
        else: #그렇지 않으면 삭제
          del bb[i]
          i-=1
        i+=1
    elif scount==3:  #3Strike
      print('**********3Strike**********')
      print('!!!!!!!!!!CORRECT!!!!!!!!!!')
      rep=False

    scount=0; bcount=0
  print('평균:',count)
  avg+=count
  count=0
  bb.clear()

print('총 평균: ', avg//1000)
```
